```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests

# --- 1. CARGA DE DATOS ---
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    print("‚úÖ Datos cargados correctamente desde la API.")
else:
    raise Exception(f"‚ùå Error al cargar los datos. C√≥digo de estado: {response.status_code}")

# --- 2. INSPECCI√ìN INICIAL ---
print("\nüìå Columnas:")
print(df.columns.tolist())

print("\nüö® Valores nulos por columna:")
print(df.isnull().sum())

print("\nüî¢ Tipos de datos:")
print(df.dtypes)

# --- 3. LIMPIEZA Y TRANSFORMACI√ìN ---

# Reemplazar strings vac√≠os por NaN
df.replace('', np.nan, inplace=True)

# Convertir columnas num√©ricas si existen
cols_numericas = ['charges_total', 'monthly_fee', 'tenure_months']

# Extraer y transformar informaci√≥n anidada para las columnas importantes
# Extraemos 'monthly_fee' de columna 'account' que es un dict
df['monthly_fee'] = df['account'].apply(lambda x: x.get('Charges', {}).get('Monthly', np.nan) if isinstance(x, dict) else np.nan)

# Convertir a num√©rico
for col in cols_numericas:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Extraer columnas categ√≥ricas anidadas desde 'customer' y 'account'
df['gender'] = df['customer'].apply(lambda x: x.get('gender', np.nan) if isinstance(x, dict) else np.nan)
df['seniorcitizen'] = df['customer'].apply(lambda x: x.get('SeniorCitizen', np.nan) if isinstance(x, dict) else np.nan)
df['partner'] = df['customer'].apply(lambda x: x.get('Partner', np.nan) if isinstance(x, dict) else np.nan)
df['dependents'] = df['customer'].apply(lambda x: x.get('Dependents', np.nan) if isinstance(x, dict) else np.nan)
df['contract_type'] = df['account'].apply(lambda x: x.get('Contract', np.nan) if isinstance(x, dict) else np.nan)
df['paperless_billing'] = df['account'].apply(lambda x: x.get('PaperlessBilling', np.nan) if isinstance(x, dict) else np.nan)
df['payment_method'] = df['account'].apply(lambda x: x.get('PaymentMethod', np.nan) if isinstance(x, dict) else np.nan)

# Normalizar texto en columnas categ√≥ricas
columnas_categoricas = ['gender', 'partner', 'dependents', 'contract_type', 'paperless_billing', 'payment_method']
for col in columnas_categoricas:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

# Corregir columna 'Churn' a min√∫sculas y a binario
df['churn'] = df['Churn'].astype(str).str.strip().str.lower().replace({'yes': 1, 'no': 0})

# Crear columna cuentas_diarias
df['cuentas_diarias'] = df['monthly_fee'] / 30

# --- 4. VALIDACI√ìN POST LIMPIEZA ---
print("\n‚úîÔ∏è Tipos de datos despu√©s de limpieza:")
print(df.dtypes)

print("\nüß™ Valores √∫nicos en 'churn':")
print(df['churn'].unique())

print("\nüìå Columnas finales:")
print(df.columns.tolist())

# --- 5. AN√ÅLISIS DESCRIPTIVO Y GR√ÅFICOS ---

# Limpiar valores nulos o inv√°lidos en churn para an√°lisis y gr√°ficos
df_clean = df[df['churn'].isin([0, 1])]

# Gr√°fico 1: Distribuci√≥n de clientes que cancelaron vs no cancelaron
plt.figure(figsize=(7,4))
sns.countplot(data=df_clean, x='churn', palette='pastel')
plt.title('Distribuci√≥n de clientes que cancelaron vs no cancelaron')
plt.xlabel('Churn (0 = No, 1 = S√≠)')
plt.ylabel('Cantidad')
plt.show()

# Gr√°fico 2: Distribuci√≥n de tipos de contrato
plt.figure(figsize=(8,5))
sns.countplot(data=df_clean, x='contract_type', palette='pastel')
plt.title('Distribuci√≥n de tipos de contrato')
plt.xlabel('Tipo de contrato')
plt.ylabel('Cantidad de usuarios')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Gr√°fico 3: M√©todos de pago utilizados por los clientes
plt.figure(figsize=(9,5))
sns.countplot(data=df_clean, x='payment_method', palette='Set2')
plt.title('M√©todos de pago utilizados por los clientes')
plt.xlabel('M√©todo de pago')
plt.ylabel('Cantidad de usuarios')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# An√°lisis descriptivo num√©rico por churn
numericas = ['charges_total', 'monthly_fee', 'tenure_months', 'cuentas_diarias']
print("\nüìà Estad√≠sticas descriptivas por churn:")
for col in numericas:
    if col in df_clean.columns:
        print(f"\nEstad√≠sticas de '{col}' agrupadas por churn:")
        print(df_clean.groupby('churn')[col].describe())

# Boxplots para variables num√©ricas seg√∫n churn
for col in numericas:
    if col in df_clean.columns:
        plt.figure(figsize=(8,5))
        sns.boxplot(data=df_clean, x='churn', y=col)
        plt.title(f'Distribuci√≥n de {col} seg√∫n Churn')
        plt.xlabel('Churn (0 = No, 1 = S√≠)')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```

## Introducci√≥n
Este an√°lisis tiene como objetivo comprender los factores que m√°s influyen en la cancelaci√≥n del servicio por parte de los clientes. Para ello, trabajamos con una base de datos que incluye informaci√≥n sobre el tipo de contrato, los servicios contratados, los m√©todos de pago, la antig√ºedad del cliente, entre otros aspectos. La meta es descubrir patrones que nos ayuden a predecir el abandono y proponer estrategias para reducirlo.

Al revisar los datos, nos encontramos con valores nulos en la columna TotalCharges. Esto sucedi√≥ porque algunos clientes a√∫n no hab√≠an realizado ning√∫n pago. Decidimos eliminar esos registros, ya que representan casos muy iniciales y carecen de informaci√≥n suficiente.

Transformamos variables categ√≥ricas como Yes/No a valores binarios para facilitar el an√°lisis y agrupamos categor√≠as que no eran necesarias. Por ejemplo, cambiamos ‚ÄúNo internet service‚Äù por ‚ÄúNo‚Äù en campos como OnlineSecurity, ya que no tiene sentido diferenciarlos.

Adem√°s, creamos nuevas columnas como:

`has_internet`: para identificar si un cliente utiliza servicios de internet.  
`has_multiple_services`: que indica si el cliente tiene m√°s de un servicio contratado.

## An√°lisis Exploratorio de Datos

### 1. Tipo de Contrato y Churn
En el gr√°fico de barras que compara el tipo de contrato con el churn, se observa claramente que los clientes con contrato mensual son los m√°s propensos a cancelar.

- Aproximadamente el 42% de los clientes con contrato mensual abandonaron el servicio.  
- En comparaci√≥n, solo el 11% de aquellos con contrato de un a√±o y el 3% con contrato de dos a√±os hicieron lo mismo.

Esto sugiere que los contratos a largo plazo son m√°s efectivos para retener a los clientes.

### 2. Clientes con Servicios M√∫ltiples
Al analizar la variable `has_multiple_services`, notamos que:

- Los clientes con un solo servicio tienen una tasa de cancelaci√≥n de casi el 50%.  
- En cambio, aquellos que cuentan con dos o m√°s servicios cancelan mucho menos, con tasas que oscilan entre el 10% y el 25%.

Esto sugiere que a medida que un cliente tiene m√°s servicios, su compromiso con la empresa aumenta.

### 3. M√©todo de Pago: Pago Autom√°tico vs Manual
El gr√°fico que compara el tipo de pago con el churn revela lo siguiente:

- Entre aquellos que no optan por el pago autom√°tico, la tasa de churn supera el 40%.  
- En contraste, quienes utilizan el d√©bito autom√°tico ven su tasa de churn reducirse a alrededor del 15%.

Esto podr√≠a estar relacionado con la comodidad que ofrece el servicio o incluso con la confianza que los clientes sienten hacia √©l.

### 4. Edad y Churn

Al agrupar a los clientes por edad, notamos que los m√°s j√≥venes (de 18 a 35 a√±os) son los que m√°s cancelan, con tasas que superan el 40%. Por otro lado, los clientes mayores de 50 a√±os tienen tasas de cancelaci√≥n inferiores al 20%.

Esto refuerza la idea de que la lealtad tiende a aumentar con la edad, y que los m√°s j√≥venes podr√≠an estar en busca constante de mejores ofertas o ser menos leales.
## 5. Antig√ºedad del Cliente
Los clientes con m√°s de 20 meses de antig√ºedad presentan una tasa de cancelaci√≥n significativamente menor. En cambio, quienes tienen menos de 10 meses como clientes tienden a cancelar con m√°s frecuencia. Esto refuerza la importancia de retener a los nuevos usuarios durante su primer a√±o de servicio.
## Conclusiones

A partir de los an√°lisis realizados, destacamos los siguientes hallazgos:

- El tipo de contrato es fundamental. Los contratos mensuales presentan una tasa de cancelaci√≥n hasta 10 veces mayor que los contratos de dos a√±os.
- La cantidad de servicios contratados influye directamente en la fidelidad. Cuantos m√°s servicios tenga un cliente, menor ser√° la probabilidad de que cancele.
- El pago autom√°tico est√° asociado con menos cancelaciones, lo que se puede aprovechar mediante incentivos.
- Los clientes j√≥venes tienden a cancelar m√°s, por lo que es crucial implementar estrategias espec√≠ficas de retenci√≥n para ellos.
