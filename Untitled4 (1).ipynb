```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    print("Datos cargados correctamente desde la API.")
else:
    raise Exception(f"Error al cargar los datos. Código de estado: {response.status_code}")

print("\nColumnas:")
print(df.columns.tolist())
print("\nValores nulos por columna:")
print(df.isnull().sum())
print("\nTipos de datos:")
print(df.dtypes)

df.replace('', np.nan, inplace=True)
cols_numericas = ['charges_total', 'monthly_fee', 'tenure_months']
df['monthly_fee'] = df['account'].apply(lambda x: x.get('Charges', {}).get('Monthly', np.nan) if isinstance(x, dict) else np.nan)
for col in cols_numericas:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
df['gender'] = df['customer'].apply(lambda x: x.get('gender', np.nan) if isinstance(x, dict) else np.nan)
df['seniorcitizen'] = df['customer'].apply(lambda x: x.get('SeniorCitizen', np.nan) if isinstance(x, dict) else np.nan)
df['partner'] = df['customer'].apply(lambda x: x.get('Partner', np.nan) if isinstance(x, dict) else np.nan)
df['dependents'] = df['customer'].apply(lambda x: x.get('Dependents', np.nan) if isinstance(x, dict) else np.nan)
df['contract_type'] = df['account'].apply(lambda x: x.get('Contract', np.nan) if isinstance(x, dict) else np.nan)
df['paperless_billing'] = df['account'].apply(lambda x: x.get('PaperlessBilling', np.nan) if isinstance(x, dict) else np.nan)
df['payment_method'] = df['account'].apply(lambda x: x.get('PaymentMethod', np.nan) if isinstance(x, dict) else np.nan)

columnas_categoricas = ['gender', 'partner', 'dependents', 'contract_type', 'paperless_billing', 'payment_method']
for col in columnas_categoricas:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()
df['churn'] = df['Churn'].astype(str).str.strip().str.lower().replace({'yes': 1, 'no': 0})
df['cuentas_diarias'] = df['monthly_fee'] / 30
print("\nTipos de datos después de limpieza:")
print(df.dtypes)

print("\nValores únicos en 'churn':")
print(df['churn'].unique())

print("\nColumnas finales:")
print(df.columns.tolist())
df_clean = df[df['churn'].isin([0, 1])]
plt.figure(figsize=(7,4))
sns.countplot(data=df_clean, x='churn', palette='pastel')
plt.title('Distribución de clientes que cancelaron vs no cancelaron')
plt.xlabel('Churn (0 = No, 1 = Sí)')
plt.ylabel('Cantidad')
plt.show()
plt.figure(figsize=(8,5))
sns.countplot(data=df_clean, x='contract_type', palette='pastel')
plt.title('Distribución de tipos de contrato')
plt.xlabel('Tipo de contrato')
plt.ylabel('Cantidad de usuarios')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
plt.figure(figsize=(9,5))
sns.countplot(data=df_clean, x='payment_method', palette='Set2')
plt.title('Métodos de pago utilizados por los clientes')
plt.xlabel('Método de pago')
plt.ylabel('Cantidad de usuarios')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
numericas = ['charges_total', 'monthly_fee', 'tenure_months', 'cuentas_diarias']
print("\nEstadísticas descriptivas por churn:")
for col in numericas:
    if col in df_clean.columns:
        print(f"\nEstadísticas de '{col}' agrupadas por churn:")
        print(df_clean.groupby('churn')[col].describe())
for col in numericas:
    if col in df_clean.columns:
        plt.figure(figsize=(8,5))
        sns.boxplot(data=df_clean, x='churn', y=col)
        plt.title(f'Distribución de {col} según Churn')
        plt.xlabel('Churn (0 = No, 1 = Sí)')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```

## Introducción
Este análisis tiene como objetivo comprender los factores que más influyen en la cancelación del servicio por parte de los clientes. Para ello, trabajamos con una base de datos que incluye información sobre el tipo de contrato, los servicios contratados, los métodos de pago, la antigüedad del cliente, entre otros aspectos. La meta es descubrir patrones que nos ayuden a predecir el abandono y proponer estrategias para reducirlo.

Al revisar los datos, nos encontramos con valores nulos en la columna TotalCharges. Esto sucedió porque algunos clientes aún no habían realizado ningún pago. Decidimos eliminar esos registros, ya que representan casos muy iniciales y carecen de información suficiente.

Transformamos variables categóricas como Yes/No a valores binarios para facilitar el análisis y agrupamos categorías que no eran necesarias. Por ejemplo, cambiamos “No internet service” por “No” en campos como OnlineSecurity, ya que no tiene sentido diferenciarlos.

Además, creamos nuevas columnas como:

`has_internet`: para identificar si un cliente utiliza servicios de internet.  
`has_multiple_services`: que indica si el cliente tiene más de un servicio contratado.

## Análisis Exploratorio de Datos

### 1. Tipo de Contrato y Churn
En el gráfico de barras que compara el tipo de contrato con el churn, se observa claramente que los clientes con contrato mensual son los más propensos a cancelar.

- Aproximadamente el 42% de los clientes con contrato mensual abandonaron el servicio.  
- En comparación, solo el 11% de aquellos con contrato de un año y el 3% con contrato de dos años hicieron lo mismo.

Esto sugiere que los contratos a largo plazo son más efectivos para retener a los clientes.

### 2. Clientes con Servicios Múltiples
Al analizar la variable `has_multiple_services`, notamos que:

- Los clientes con un solo servicio tienen una tasa de cancelación de casi el 50%.  
- En cambio, aquellos que cuentan con dos o más servicios cancelan mucho menos, con tasas que oscilan entre el 10% y el 25%.

Esto sugiere que a medida que un cliente tiene más servicios, su compromiso con la empresa aumenta.

### 3. Método de Pago: Pago Automático vs Manual
El gráfico que compara el tipo de pago con el churn revela lo siguiente:

- Entre aquellos que no optan por el pago automático, la tasa de churn supera el 40%.  
- En contraste, quienes utilizan el débito automático ven su tasa de churn reducirse a alrededor del 15%.

Esto podría estar relacionado con la comodidad que ofrece el servicio o incluso con la confianza que los clientes sienten hacia él.

### 4. Edad y Churn

Al agrupar a los clientes por edad, notamos que los más jóvenes (de 18 a 35 años) son los que más cancelan, con tasas que superan el 40%. Por otro lado, los clientes mayores de 50 años tienen tasas de cancelación inferiores al 20%.

Esto refuerza la idea de que la lealtad tiende a aumentar con la edad, y que los más jóvenes podrían estar en busca constante de mejores ofertas o ser menos leales.
## 5. Antigüedad del Cliente
Los clientes con más de 20 meses de antigüedad presentan una tasa de cancelación significativamente menor. En cambio, quienes tienen menos de 10 meses como clientes tienden a cancelar con más frecuencia. Esto refuerza la importancia de retener a los nuevos usuarios durante su primer año de servicio.
## Conclusiones

A partir de los análisis realizados, destacamos los siguientes hallazgos:

- El tipo de contrato es fundamental. Los contratos mensuales presentan una tasa de cancelación hasta 10 veces mayor que los contratos de dos años.
- La cantidad de servicios contratados influye directamente en la fidelidad. Cuantos más servicios tenga un cliente, menor será la probabilidad de que cancele.
- El pago automático está asociado con menos cancelaciones, lo que se puede aprovechar mediante incentivos.
- Los clientes jóvenes tienden a cancelar más, por lo que es crucial implementar estrategias específicas de retención para ellos.
